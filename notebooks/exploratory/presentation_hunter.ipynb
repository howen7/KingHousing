```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df['Porch_y_n'] = np.where(df['SqFtOpenPorch'].apply(lambda x: x > 0), 'Porch', 'No Porch')
```

```python
df['Porch_y_n'] = np.where(df['SqFtOpenPorch'].apply(lambda x: x > 0), 'Porch', 'No Porch')
```

```python
Porch_sales = np.array(df.SalePrice.loc[df.Porch_y_n == 'Porch'])
no_porch_sales = np.array(df.SalePrice.loc[df.Porch_y_n == 'No Porch'])
```

```python

fig, ax = plt.subplots(figsize=(17,10))
ax = sns.barplot(x =df['Porch_y_n'], y = df['SalePrice'], palette = "Blues_d")
sns.set_context(context= 'poster', font_scale = 2)
ax.set(xlabel = '', ylabel = "Home Selling Price")
ax.legend()
plt.show()
```

```python
fig.savefig('../../figures/Porch.png',dpi=300, bbox_inches='tight') 
```

```python

fig, ax = plt.subplots(figsize=(17,10))
ax = sns.distplot(Porch_sales, hist = True, bins = 100)
ax = sns.distplot(no_porch_sales, hist = True, bins = 100)
sns.set_context(context= 'poster', font_scale = 2)
ax.set(xlabel = '', ylabel = "Home Selling Price")
ax.legend()
plt.show()
```
