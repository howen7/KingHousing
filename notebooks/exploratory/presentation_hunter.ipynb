```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../../src/data/combined_onehot.json')
```

```python
df['Porch_y_n'] = np.where(df['SqFtOpenPorch'].apply(lambda x: x > 0), 'Porch', 'No Porch')
```

```python
Porch_sales = np.array(df.SalePrice.loc[df.Porch_y_n == 'Porch'])
no_porch_sales = np.array(df.SalePrice.loc[df.Porch_y_n == 'No Porch'])
```

```python

fig, ax = plt.subplots(figsize=(17,10))
ax = sns.barplot(x =df['Porch_y_n'], y = df['SalePrice'], palette = "Blues_d", )
sns.set_context(context= 'poster', font_scale = 2)
ax.set(xlabel = '', ylabel = "Home Selling Price",title = 'Porch VS No Porch Mean Home Sale Prices')
plt.show()
fig.savefig('../report/figures/Porch.png',dpi=300, bbox_inches='tight') 
```
